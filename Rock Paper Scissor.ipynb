```python
def result(h,c):#function returning the winner "h" for human and "c" for computer 
    choice=("P","R","S")
    if h=="P":
        if c=="R":
            return "h"
        elif c=="S":
            return "c"
        else:
            return "Tie"
    elif h=="R":
        if c=="S":
            return "h"
        elif c=="P":
            return "c"
        else:
            return "Tie"
    elif h=="S":
        if c=="P":
            return "h"
        elif c=="R":
            return "c"
        else:
            return "Tie"

        

def fin(x,y):#displaying end the result
    if x>y:
        print(f"Player name = {score_h} | Robot = {score_c}\n")
        print("Congrats! You Won")
    elif x<y:
        print(f"Player name = {score_h} | Robot = {score_c}\n")
        print("Sorry! You Lose, Try again")
    else:
        print(f"Player name = {score_h} | Robot = {score_c}\n")
        print("Match Tie, Try again")

    

import random
import time
cool=("ready","steady","Go!")
score_h,score_c=0,0#assigning variable to store score for player and computer
name=input("Enter your name: ")
print("Do you want to play 'Rock, Paper and Scissor'")
while True:
    x=0
    q=input("Press [Y] to play or [N] to Quit: ").upper()
    if q.strip()=="Y":
        print("Awesome!")
        y=True
        while y:
            try:
                rounds=int(input("Enter the number of rounds: "))
                y=False
            except:
                print("Enter valid argument")
        while x<rounds:
            for l in range(0,len(cool)):
                print(cool[l])
                time.sleep(1)
            print()
            user_action = input("Enter a choice (rock[R], paper[P], scissors[S]): ").upper()
            possible_actions = ["R", "P", "S"]
            computer = random.choice(possible_actions)
            if result(user_action.strip(),computer) == "h":
                print(f"\nYou chose {user_action}, computer chose {computer}.\n")
                print(f"{name} won, {name} got 1 point")
                score_h+=1
                print(f"Player name = {score_h} | Robot = {score_c}\n")
                x+=1
            elif result(user_action.strip(),computer) == "c":
                print(f"\nYou chose {user_action}, computer chose {computer}.\n")
                print(f"{name} lose, Computer got 1 point")
                score_c+=1
                print(f"Player name = {score_h} | Robot = {score_c}\n")
                x+=1
            elif result(user_action.strip(),computer) == "Tie":
                print(f"\nYou chose {user_action}, computer chose {computer}.\n")
                print("It's a tie game.\n no body wins")
                print(f"Player name = {score_h} | Robot = {score_c}\n")
                x+=1
            else:
                print("Enter a valid argument")
        print("\n\nGame Over! Final score.\n")
        fin(score_h,score_c)
        print("Do you like to play again")
    elif q.strip()=="N":
        print("That's cool, have a good one!")
        break    
    else:
        print("Enter a valid argument")


        

```
