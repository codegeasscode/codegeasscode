```python
from tkinter import*
calculator_screen = Tk()

screen=Entry(calculator_screen)
screen.grid(row=0,column=0,columnspan=4)

global a
a=False
def click(num):
    global a
    if num in ['+','-','*','/']:
        if a==True:
            equal()
            a=False
        else:
            a=True
    
    current=screen.get()
    screen.delete(0,END)
    screen.insert(0,current+str(num))
    
    
    
def equal():
    calc=eval(screen.get())
    screen.delete(0,END)
    screen.insert(0,calc)

    
def clear():
    screen.delete(0,END)
    



button1=Button(calculator_screen,text="1",command=lambda: click(1))
button2=Button(calculator_screen,text="2",command=lambda: click(2))
button3=Button(calculator_screen,text="3",command=lambda: click(3))

button4=Button(calculator_screen,text="4",command=lambda: click(4))
button5=Button(calculator_screen,text="5",command=lambda: click(5))
button6=Button(calculator_screen,text="6",command=lambda: click(6))

button7=Button(calculator_screen,text="7",command=lambda: click(7))
button8=Button(calculator_screen,text="8",command=lambda: click(8))
button9=Button(calculator_screen,text="9",command=lambda: click(9))

button0=Button(calculator_screen,text="0",command=lambda: click(0))
button_add=Button(calculator_screen,text="+",command=lambda:click('+'))
button_subtract=Button(calculator_screen,text="-",command=lambda:click('-'))

button_multiply=Button(calculator_screen,text="*",command=lambda:click('*'))
button_division=Button(calculator_screen,text="/",command=lambda:click('/'))
button_equal=Button(calculator_screen,text="=",command=equal)

button_dot=Button(calculator_screen,text='.',command=lambda:click('.'))
button_clear=Button(calculator_screen,text="clear",command= clear)

button1.grid(row=4,column=0)
button2.grid(row=4,column=1)
button3.grid(row=4,column=2)

button4.grid(row=3,column=0)
button5.grid(row=3,column=1)
button6.grid(row=3,column=2)

button7.grid(row=2,column=0)
button8.grid(row=2,column=1)
button9.grid(row=2,column=2)

button0.grid(row=5,column=0)
button_add.grid(row=5,column=3)
button_subtract.grid(row=3,column=3)

button_multiply.grid(row=4,column=3)
button_division.grid(row=5,column=1)
button_equal.grid(row=5,column=2,columnspan=2)

button_dot.grid(row=6,column=1,columnspan=2)
button_clear.grid(row=6,column=0,columnspan=2)

calculator_screen.mainloop()
```
