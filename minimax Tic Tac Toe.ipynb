```python
from tkinter import *
from tkinter import messagebox
window=Tk()
screen=Label(window,text='Player:x          Computer:o',font='Helvetica',fg='Red',bg='Orange')
screen.grid(row=0,column=0,columnspan=3)
player, opponent = 'o', 'x'
board = [
    ['_','_','_'],
    ['_','_','_'],
    ['_','_','_']
]
global count
count=0
def isMovesLeft(board) :

    for i in range(3) :
        for j in range(3) :
            if (board[i][j] == '_') :
                return True
    return False

def evaluate(b) :

    for row in range(3) :	
        if (b[row][0] == b[row][1] and b[row][1] == b[row][2]) :	
            if (b[row][0] == player) :
                return 10
            elif (b[row][0] == opponent) :
                return -10
    for col in range(3) :
        if (b[0][col] == b[1][col] and b[1][col] == b[2][col]) :
            if (b[0][col] == player) :
                return 10
            elif (b[0][col] == opponent) :
                return -10

    if (b[0][0] == b[1][1] and b[1][1] == b[2][2]) :
        if (b[0][0] == player) :
            return 10
        elif (b[0][0] == opponent) :
            return -10

    if (b[0][2] == b[1][1] and b[1][1] == b[2][0]) :
        if (b[0][2] == player) :
            return 10
        elif (b[0][2] == opponent) :
            return -10
    return 0

def minimax(board, depth, isMax) :
    score = evaluate(board)
    if (score == 10) :
        return score
    if (score == -10) :
        return score
    if (isMovesLeft(board) == False) :
        return 0
    if (isMax) :
        best = -1000
        for i in range(3) :		
            for j in range(3) :
                if (board[i][j]=='_') :
                    board[i][j] = player
                    best = max( best, minimax(board,depth + 1,not isMax) )
                    board[i][j] = '_'
        return best
    else :
        best = 1000
        for i in range(3) :
            for j in range(3) :
                if (board[i][j] == '_') :
                    board[i][j] = opponent
                    best = min(best, minimax(board, depth + 1, not isMax))
                    board[i][j] = '_'
        return best

def findBestMove(board) :
    bestVal = -1000
    bestMove = (-1, -1)
    for i in range(3) :	
        for j in range(3) :
            if (board[i][j] == '_') :
                board[i][j] = player
                moveVal = minimax(board, 0, False)
                board[i][j] = '_'
                if (moveVal > bestVal) :			
                    bestMove = (i, j)
                    bestVal = moveVal

    return bestMove

            




        
    
def click(a,b):
    global count
    button=[[b1,b2,b3],[b4,b5,b6],[b7,b8,b9]]
    if button[a][b]['text'] == '':
        button[a][b]['text']= 'x'
        board[a][b]='x'
        best_move=findBestMove(board)
        button[best_move[0]][best_move[1]]['text']='o'
        board[best_move[0]][best_move[1]]='o'
        count+=2
        result=evaluate(board)
        if result == 10:
            messagebox.showinfo('result','Computer wins!, You Lose')
            count=0
            for i in range(3):
                for j in range(3):
                    button[i][j]['text']=''
                    board[i][j]='_'
        elif result == -10:
            messagebox.showinfo('result','Great! You win')
            count=0
            for i in range(3):
                for j in range(3):
                    button[i][j]['text']=''
                    board[i][j]='_'
            
        elif result == 0 and count==10:
            messagebox.showinfo('result',"Nobody wins!, It's a tie")
            count=0
            for i in range(3):
                for j in range(3):
                    button[i][j]['text']=''
                    board[i][j]='_'

        

b1= Button(window,command=lambda: click(0,0),height=10,width=10,bg='olive',font=('Helvetica',10))
b2= Button(window,command=lambda: click(0,1),height=10,width=10,bg='olive',font=('Helvetica',10))
b3= Button(window,command=lambda: click(0,2),height=10,width=10,bg='olive',font=('Helvetica',10))
b4= Button(window,command=lambda: click(1,0),height=10,width=10,bg='olive',font=('Helvetica',10))
b5= Button(window,command=lambda: click(1,1),height=10,width=10,bg='olive',font=('Helvetica',10))
b6= Button(window,command=lambda: click(1,2),height=10,width=10,bg='olive',font=('Helvetica',10))
b7= Button(window,command=lambda: click(2,0),height=10,width=10,bg='olive',font=('Helvetica',10))
b8= Button(window,command=lambda: click(2,1),height=10,width=10,bg='olive',font=('Helvetica',10))
b9= Button(window,command=lambda: click(2,2),height=10,width=10,bg='olive',font=('Helvetica',10))

b1.grid(row=1,column=0)
b2.grid(row=1,column=1)
b3.grid(row=1,column=2)
b4.grid(row=2,column=0)
b5.grid(row=2,column=1)
b6.grid(row=2,column=2)
b7.grid(row=3,column=0)
b8.grid(row=3,column=1)
b9.grid(row=3,column=2)

window.mainloop()
```
