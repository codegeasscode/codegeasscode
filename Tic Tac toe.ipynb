```python
# importing tkinter and messagebox
from tkinter import *
from tkinter import messagebox
# making root window
window=Tk()
# setting size of window
window.minsize(100,300)
# making label in window
screen=Label(window,text='Player1:x          Player2:o',font='Helvetica',fg='Red',bg='Orange')
screen.grid(row=0,column=0,columnspan=3)
# assigning symbols to each player
p1='x'
p2='o'
# setting count =0 for initial 
global count
count=0
# check() function to check who is winner
def check():
    global count
    button={1:b1,2:b2,3:b3,4:b4,5:b5,6:b6,7:b7,8:b8,9:b9}
    if (b1['text'] == "o" and b2['text'] == "o" and b3['text'] == "o") or (b4['text'] == "o" and b5['text'] == "o" and b6['text'] == "o") or (b7['text'] == "o" and b8['text'] == "o" and b9['text'] == "o") or (b1['text'] == "o" and b4['text'] == "o" and b7['text'] == "o") or (b2['text'] == "o" and b5['text'] == "o" and b8['text'] == "o") or (b3['text'] == "o" and b6['text'] == "o" and b9['text'] == "o") or (b1['text'] == "o" and b5['text'] == "o" and b9['text'] == "o") or (b3['text'] == "o" and b5['text'] == "o" and b7['text'] == "o"):
        messagebox.showinfo("Winner", "Player2 wins!")
        for i in range(1,10):
            button[i]['text']=''
        count=0
    elif (b1['text'] == "x" and b2['text'] == "x" and b3['text'] == "x") or (b4['text'] == "x" and b5['text'] == "x" and b6['text'] == "x") or (b7['text'] == "x" and b8['text'] == "x" and b9['text'] == "x") or (b1['text'] == "x" and b4['text'] == "x" and b7['text'] == "x") or (b2['text'] == "x" and b5['text'] == "x" and b8['text'] == "x") or (b3['text'] == "x" and b6['text'] == "x" and b9['text'] == "x") or (b1['text'] == "x" and b5['text'] == "x" and b9['text'] == "x") or (b3['text'] == "x" and b5['text'] == "x" and b7['text'] == "x"): 
        messagebox.showinfo("Winner", "Player1 wins!")
        for i in range(1,10):
            button[i]['text']=''
        count=0
    elif b1['text'] !='' and b2['text'] !='' and b3['text'] !='' and b4['text'] !='' and b5['text'] !='' and b6['text'] !='' and b7['text'] !='' and b8['text'] !='' and b9['text'] !='':
        messagebox.showinfo("Winner", "Tie!")
        for i in range(1,10):
            button[i]['text']=''
        count=0



# toggling the position for each player over window
def choose():
    global count
    if count%2==0:
        return 'x'
    elif count%2 !=0:
        return 'o'
    

# intializing the value to buttons
def click(a):
    global count
    button={1:b1,2:b2,3:b3,4:b4,5:b5,6:b6,7:b7,8:b8,9:b9}
    if button[a]['text'] == '':
        button[a]['text']= choose()
        count+=1
    check()
         
        
# making 9 buttons for each position
b1= Button(window,command=lambda: click(1),height=10,width=10,bg='olive')
b2= Button(window,command=lambda: click(2),height=10,width=10,bg='olive')
b3= Button(window,command=lambda: click(3),height=10,width=10,bg='olive')
b4= Button(window,command=lambda: click(4),height=10,width=10,bg='olive')
b5= Button(window,command=lambda: click(5),height=10,width=10,bg='olive')
b6= Button(window,command=lambda: click(6),height=10,width=10,bg='olive')
b7= Button(window,command=lambda: click(7),height=10,width=10,bg='olive')
b8= Button(window,command=lambda: click(8),height=10,width=10,bg='olive')
b9= Button(window,command=lambda: click(9),height=10,width=10,bg='olive')


b1.grid(row=1,column=0)
b2.grid(row=1,column=1)
b3.grid(row=1,column=2)
b4.grid(row=2,column=0)
b5.grid(row=2,column=1)
b6.grid(row=2,column=2)
b7.grid(row=3,column=0)
b8.grid(row=3,column=1)
b9.grid(row=3,column=2)

# running mainloop
window.mainloop()
```
