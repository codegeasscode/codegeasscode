```python
from tkinter import*
calculator_screen = Tk()

screen=Entry(calculator_screen)
screen.grid(row=0,column=0,columnspan=4)

def click(num):
    current=screen.get()
    screen.delete(0,END)
    screen.insert(0,current+str(num))
    
    
def add():
    global first_num
    first_num=int(screen.get())
    screen.delete(0,END)
    global sign
    sign="+"


def subtract():
    global first_num
    first_num=int(screen.get())
    screen.delete(0,END)
    global sign
    sign="-"



def multiply():
    global first_num
    first_num=int(screen.get())
    screen.delete(0,END)
    global sign
    sign="*"



def divide():
    global first_num
    first_num=int(screen.get())
    screen.delete(0,END)
    global sign
    sign="/"



def equal():
    second_num=int(screen.get())
    screen.delete(0,END)
    if sign == "+":
        screen.insert(0,first_num+second_num)
    elif sign == "-":
        screen.insert(0,first_num-second_num)
    elif sign == "*":
        screen.insert(0,first_num*second_num)
    else:
        if second_num !=0:
            screen.insert(0,first_num/second_num)
        else:
            screen.delete(0,END)


    
def clear():
    screen.delete(0,END)
    


button1=Button(calculator_screen,text="1",command=lambda: click(1))
button2=Button(calculator_screen,text="2",command=lambda: click(2))
button3=Button(calculator_screen,text="3",command=lambda: click(3))

button4=Button(calculator_screen,text="4",command=lambda: click(4))
button5=Button(calculator_screen,text="5",command=lambda: click(5))
button6=Button(calculator_screen,text="6",command=lambda: click(6))

button7=Button(calculator_screen,text="7",command=lambda: click(7))
button8=Button(calculator_screen,text="8",command=lambda: click(8))
button9=Button(calculator_screen,text="9",command=lambda: click(9))

button0=Button(calculator_screen,text="0",command=lambda: click(0))
button_add=Button(calculator_screen,text="+",command=add)
button_subtract=Button(calculator_screen,text="-",command=subtract)

button_multiply=Button(calculator_screen,text="*",command=multiply)
button_division=Button(calculator_screen,text="/",command=divide)
button_equal=Button(calculator_screen,text="=",command=equal)

button_clear=Button(calculator_screen,text="clear",command= clear)

button1.grid(row=3,column=0)
button2.grid(row=3,column=1)
button3.grid(row=3,column=2)

button4.grid(row=2,column=0)
button5.grid(row=2,column=1)
button6.grid(row=2,column=2)

button7.grid(row=1,column=0)
button8.grid(row=1,column=1)
button9.grid(row=1,column=2)

button0.grid(row=4,column=0)
button_add.grid(row=1,column=3)
button_subtract.grid(row=2,column=3)

button_multiply.grid(row=3,column=3)
button_division.grid(row=4,column=1)
button_equal.grid(row=4,column=2,columnspan=2)

button_clear.grid(row=5,column=0,columnspan=4)

calculator_screen.mainloop()
```
